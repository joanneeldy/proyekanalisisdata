2. Data Cleaning & Preprocessing
Tujuan:
Membersihkan data dari ketidaksesuaian, duplikasi, nilai hilang, dan kolom-kolom yang tidak relevan, agar data siap digunakan untuk analisis lebih lanjut.

Langkah-langkah yang dilakukan:

- Konversi Tanggal: Mengubah kolom dteday menjadi format datetime untuk memudahkan analisis berbasis waktu.
- Menghapus Kolom Redundan:
    -> Menghapus kolom instant (hanya sebagai nomor urut) yang tidak memberikan informasi analitis.
    -> Menghapus kolom workingday jika informasi ini sudah tergantikan oleh kolom weekday.
- Merging Data:
    -> Menggabungkan data dari day.csv dan hour.csv untuk mendapatkan satu tabel lengkap.
    -> Misalnya, menambahkan kolom daily_cnt (total penyewaan harian) dari data harian ke data per jam.
- Scaling Nilai:
    -> Data seperti temp, atemp, humidity, dan windspeed awalnya ternormalisasi. Langkah ini mengalikan nilai dengan konstanta untuk mengembalikannya ke skala aslinya (misal: suhu dalam °C) agar lebih mudah diinterpretasikan.
- Konversi Kode Kategorikal:
    -> Mengubah kode numerik di kolom seperti season, month, weather_category, weekday, dan year menjadi label yang deskriptif (misalnya, 1 → Spring, 2 → Summer, dll.).
- Penambahan Kolom Kategori:
    -> Menambahkan kolom category_days untuk mengklasifikasikan hari sebagai weekend atau weekdays.
    -> Menambahkan kolom humidity_category untuk mengkategorikan kelembapan menjadi “Terlalu kering”, “Ideal”, dan “Terlalu Lembab”.

Insight dari Data Cleaning:

- Data yang sudah bersih dan konsisten memungkinkan analisis lebih mendalam.
- Informasi seperti nilai suhu dan kelembapan sekarang berada di skala asli, sehingga hubungan dengan jumlah penyewaan bisa lebih mudah diinterpretasikan.
- Penggabungan data dari dua file menjadi satu tabel lengkap memudahkan eksplorasi pola per jam dengan konteks agregat harian.

```python
# change the 'dteday' column to datetime format in day_df & hour_df
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])
```

```python
# merge data

# make a subset from day_df with dteday & cnt
day_subset = day_df[['dteday', 'cnt']].copy()

# change 'cnt' to 'daily_cnt' so it's not conflicted with cnt in hour_df
day_subset.rename(columns={'cnt': 'daily_cnt'}, inplace=True)

# merge hour_df with day_subset by 'dteday'
merged_df = pd.merge(hour_df, day_subset, on='dteday', how='left')

# save merged data
merged_df.to_csv('Bike-sharing-dataset/merged.csv', index=False)

# show merged data
print("Data hasil merge:")
print(merged_df.head())
```

```python
# make the normalized value into its normal scale
merged_df['temp'] = merged_df['temp'] * 41
merged_df['atemp'] = merged_df['atemp'] * 50
merged_df['humidity'] = merged_df['humidity'] * 100
merged_df['windspeed'] = merged_df['windspeed'] * 67
```

```python
# convert the categorical value into descriptive label

# convert season (1:Spring, 2:Summer, 3:Fall, 4:Winter)
merged_df['season'] = merged_df['season'].replace({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})

# convert month (1:Jan, 2:Feb, ..., 12:Dec)
merged_df['month'] = merged_df['month'].replace({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})

# convert weather_category (1-4)
merged_df['weather_category'] = merged_df['weather_category'].replace({
    1: 'Clear', 2: 'Misty', 3: 'Light_rainsnow', 4: 'Heavy_rainsnow'
})

# convert weekday (numbers to name of days)
merged_df['weekday'] = merged_df['weekday'].replace({
    0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 
    4: 'Thursday', 5: 'Friday', 6: 'Saturday'
})

# convert year (0:2011, 1:2012)
merged_df['year'] = merged_df['year'].replace({0: '2011', 1: '2012'})
```

```python
# make  category_days: mengklasifikasikan hari sebagai weekend atau weekdays
def get_category_days(day_name):
    if day_name in ['Saturday', 'Sunday']:
        return 'weekend'
    else:
        return 'weekdays'

merged_df['category_days'] = merged_df['weekday'].apply(get_category_days)

# Fungsi untuk mengklasifikasi kelembapan
def classify_humidity(humidity):
    if humidity < 45:
        return "Terlalu kering"
    elif 45 <= humidity < 65:
        return "Ideal"
    else:
        return "Terlalu Lembab"

merged_df['humidity_category'] = merged_df['humidity'].apply(classify_humidity)
```

```python
# rename columns
merged_df.rename(columns={
    "dteday": "dateday",
    "yr": "year",
    "hr": "hour",
    "mnth": "month",
    "weathersit": "weather_category",
    "hum": "humidity",
    "cnt": "hourly_count"  # cnt from hour.csv (total rents per hour)
}, inplace=True)
```

```python
# delete column: 'workingday'
merged_df.drop(columns=['instant', 'workingday'], errors='ignore', inplace=True)
```

```python
merged_df.to_csv('Bike-sharing-dataset/clean_merged.csv', index=False)
```

3. Exploratory Data Analysis (EDA)
Tujuan:
Menjelajahi data secara visual dan statistik untuk menemukan pola, tren, serta hubungan antar variabel. Hasil analisis ini akan menjadi dasar dalam menjawab pertanyaan bisnis yang telah ditetapkan.

Pertanyaan Bisnis yang Dijawab:

Bagaimana tren penyewaan sepeda berubah seiring waktu?

Visualisasi: Line plot yang menggambarkan tren total penyewaan harian (daily_cnt) terhadap tanggal (dateday).
Insight: Dapat dilihat adanya pola musiman, misalnya peningkatan penyewaan pada musim panas dan penurunan pada musim dingin.
Bagaimana pengaruh faktor waktu (jam) terhadap penyewaan?

Visualisasi: Bar plot rata-rata penyewaan per jam (hourly_count) berdasarkan kolom hr.
Insight: Menunjukkan jam-jam tertentu (misalnya pagi dan sore) yang merupakan jam sibuk dengan penyewaan tertinggi.
Bagaimana pengaruh kondisi musim (season) terhadap penyewaan?

Visualisasi: Box plot yang memperlihatkan distribusi penyewaan per jam berdasarkan musim (season).
Insight: Terlihat perbedaan yang signifikan antara musim, yang dapat membantu dalam perencanaan operasional.
Bagaimana pengaruh kelembapan terhadap penyewaan?

Visualisasi: Bar plot rata-rata penyewaan berdasarkan kategori kelembapan (humidity_category).
Insight: Menunjukkan kondisi cuaca optimal (misalnya, kelembapan ideal) yang mendukung peningkatan penyewaan.
Apakah terdapat perbedaan pola antara hari kerja (weekdays) dan akhir pekan (weekend)?

Visualisasi: Box plot perbandingan penyewaan antara weekdays dan weekend (kolom category_days).
Insight: Bisa terlihat apakah penyewaan lebih tinggi pada hari kerja atau akhir pekan.

```python
df = pd.read_csv('Bike-sharing-dataset/clean_merged.csv')

print(df.head())
```

```python
df.info()

print(df.describe())
```

```python
# total rents per day graph
plt.figure()
sns.lineplot(x=df['dateday'], y=df['daily_cnt'])
plt.title('Tren Penyewaan Sepeda per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Total Penyewaan')
plt.xticks(rotation=45)
plt.show()
```

```python
# average rents per hour
avg_by_hour = df.groupby('hour')['hourly_count'].mean().reset_index()

plt.figure()
sns.barplot(x=avg_by_hour['hour'], y=avg_by_hour['hourly_count'])
plt.title('Rata-rata Penyewaan Sepeda Berdasarkan Jam')
plt.xlabel('Jam')
plt.ylabel('Rata-rata Penyewaan')
plt.show()
```

```python
# how season affects rent
plt.figure()
sns.boxplot(x=df['season'], y=df['hourly_count'])
plt.title('Distribusi Penyewaan per Jam Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Penyewaan per Jam')
plt.show()
```

```python
# how weather affects rent
plt.figure()
sns.barplot(x=df['weather_category'], y=df['hourly_count'])
plt.title('Rata-rata Penyewaan per Jam Berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Rata-rata Penyewaan')
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['category_days'], y=df['hourly_count'])
plt.title('Perbandingan Penyewaan pada Weekdays vs Weekend')
plt.xlabel('Kategori Hari')
plt.ylabel('Penyewaan per Jam')
plt.show()
```

```python
# temperature distribution
plt.figure()
sns.histplot(df['temp'], bins=30, kde=True)
plt.title('Distribusi Suhu')
plt.xlabel('Suhu (°C)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# temperature & rent relationship
plt.figure()
sns.scatterplot(x=df['temp'], y=df['hourly_count'])
plt.title('Hubungan Suhu dengan Penyewaan Sepeda per Jam')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penyewaan per Jam')
plt.show()
```

```python
# humidity x hourly count
plt.figure()
sns.barplot(x=df['humidity'], y=df['hourly_count'])
plt.title('Rata-rata Penyewaan berdasarkan Kategori Kelembapan')
plt.xlabel('Kategori Kelembapan')
plt.ylabel('Rata-rata Penyewaan')
plt.show()
```
